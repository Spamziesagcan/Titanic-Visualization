```python
numeric_df = combined_df.select_dtypes(include=['number'])
```

```python
corr_matrix = numeric_df.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".2f")
plt.title("Overall Correlation Heatmap", fontsize=14)
plt.show()
```

***Columns like Pclass, Survived, Fare and Age are heavily correlated to each other, i.e., their values are affected by change in any of those values.***

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Sex', y='Age', data=combined_df, palette='coolwarm')
plt.title("Distribution of Sex by Age", fontsize=14)
plt.show()
```

***Both male and female passengers had a median age range 22-26. It looks like female had a higher number of passengers who were 40 or above as compared to male.***

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Sex', y='Pclass', data=combined_df, palette='coolwarm')
plt.title("Distribution of Sex by Class", fontsize=14)
plt.show()
```

***This depicts that the total number of female passengers was more than male passengers as the plot is denser for female passengers at every class.***

```python
plt.figure(figsize=(6, 14))
sns.boxplot(x='Sex', y='Fare', data=combined_df, palette='coolwarm')
plt.title("Box Plot of Fares across Sex", fontsize=14)
plt.show()
```

***This depicts that both male and female passengers acquired the class 3, nevertheless there were more female passengers than male passengers who acquired class 1.***

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='Age', data=combined_df, palette='coolwarm')
plt.title("Box Plot of Age across Sex", fontsize=14)
plt.show()
```

***This depicts that both male and female passengers had the same average age. It is seen that most of the old passengers (70+) were male.***

```python
plt.figure(figsize=(6, 6))
sns.histplot(combined_df['Pclass'], kde=True, color='skyblue', bins=20)
plt.title("Histogram of Class", fontsize=14)
plt.xlabel("Class")
plt.ylabel("Frequency")
plt.show()
```

***This depicts that class 1 was the most acquired class on the ship, which tells us that the ship was full of rich people.***

```python
plt.figure(figsize=(10, 6))
sns.histplot(combined_df['Sex'], kde=True, color='skyblue', bins=20)
plt.title("Histogram of Sex", fontsize=14)
plt.xlabel("Sex")
plt.ylabel("Frequency")
plt.show()
```

***This depicts the comparison between the count of male and female passengers, i.e., there were more male passengers than female passengers on the ship.***

```python
plt.figure(figsize=(10, 6))
sns.histplot(combined_df['Age'], kde=True, color='skyblue', bins=20)
plt.title("Histogram of Age", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

***This depicts the count of passengers based on their age. It looks like majority of the passengers aged 20-35.***

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=combined_df, palette='coolwarm', s=100)
plt.title("Scatter Plot of Age vs Fare", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

***This depicts the comparison of passengers of different ages and which class they preferred or could afford which can be judged based on the fare they paid.***
